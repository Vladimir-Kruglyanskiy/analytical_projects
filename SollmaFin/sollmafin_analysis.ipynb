#  Исследование результатов А/В-теста и поиск инсайтов

SollmaFin - международное мобильное приложение для онлайн-торговли финансовыми активами. Пользователи могут инвестировать деньги в акции, валюту, криптовалюту, биржевые фонды (ETF) и другие активы. Целевая аудитория приложения — начинающие инвесторы. Получить доступ к финансовым рынкам можно со смартфона либо через веб-версию. 

Приложение ориентировано на рынок Латинской Америки с акцентом на четыре страны: Мексику, Бразилию, Колумбию и Аргентину - там быстро растёт интерес к финтеху и инвестициям, а экономики очень разные. 

Появилась гипотеза о недостаточной финансовой грамотности пользователей: они не понимают, как работать с активами разной степени риска. Клиенты покупают активы с высоким риском и теряют деньги, что снижает вовлечённость.
Команда продукта решила обновить онбординг пользователей и добавить в него детальную информацию о различиях финансовых активов и связанных с ними рисках. Однако появилось опасение, что углублённый онбординг отпугнёт клиентов от пополнения депозита и более рискованных вложений. 

Чтобы разобраться, команда решила провести A/B-эксперимент. 

### А/В-эксперимент

В эксперименте участвовали новые пользователи, которые зарегистрировались в приложении со 2 по 15 июня 2025 года. 
Пользователей случайным образом разделили на две равные группы: 
- Контрольная группа проходила стандартный онбординг без обязательного обучения.
- Тестовая группа проходила обновлённый онбординг с подробной информацией об активах и связанных с ними рисках. 

После разделения активность пользователей анализировали в течение недели.

Цель эксперимента - оценить, как обновлённый онбординг влияет на поведение пользователей и их дальнейшую инвестиционную активность.

При внедрении новой фичи команда продукта выдвинула такие гипотезы:
- Гипотеза роста: обучающий онбординг помогает пользователям лучше понимать принципы инвестирования, поэтому они будут чаще открывать второй депозит.
- Гипотеза риска: информация о возможных потерях и высоких рисках отпугнёт некоторых новичков, особенно самых осторожных, что снизит конверсию в первый депозит.
- Дополнительная гипотеза: после нового онбординга пользователи, которые выбрали высокорискованные активы, будут чаще, чем раньше, возвращаться и открывать второй депозит. При старом онбординге пользователи часто покупали активы с высоким риском без понимания последствий. Это приводило к потерям и оттоку после первого депозита. 

Команда выбрала для эксперимента такой набор метрик: 
- Ключевая метрика — средняя сумма всех депозитов на одного пользователя (включая тех, кто установил приложение или открыл веб-версию).
- Барьерная метрика — конверсия из регистрации в первый депозит.
- Вспомогательная метрика 1 — конверсия из первого депозита во второй.
- Вспомогательная метрика 2 — средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит.

2 датасета:  

1. Датасет `/datasets/pa_sollmafin_hist.csv` содержит исторические данные о ключевых действиях новых пользователей, привлечённых в период с 1 апреля по 1 июня 2025 года включительно. В датасете собраны действия пользователей до оформления второго депозита. 
2. Датасет `/datasets/pa_sollmafin_abt.csv` содержит данные А/В-эксперимента - все действия новых пользователей, которые зарегистрировались со 2 по 15 июня 2025 года включительно. Общие поля датасетов:

- `asset` — тип приобретённого актива; 
- `risk_level` — уровень риска актива: low — низкий риск, medium — средний риск, high — высокий риск.

В датасете `/datasets/pa_sollmafin_abt.csv` содержатся два дополнительных поля:
- `ab_test` — название А/В-эксперимента;

**Анализ воронок событий.**


```python
# df_hist.head()
```

Из шагов воронки убираем `introduction`, т. к. кол-во пользователей на этом шаге равно кол-ву пользователей на шаге `install / open_web`

```python

steps = ['install / open_web', 'registration', 'main_page', 'onboarding_complete', 'first_deposit', 'asset_purchase', 'second_deposit']

df_funnel = df_hist.groupby('event_name')['user_id'].nunique().reindex(steps).reset_index()
df_funnel['cr(%)'] = round(df_funnel['user_id'] / df_funnel['user_id'].iloc[0] * 100, 2)

df_funnel['step_by_step_cr'] = round(df_funnel['user_id'] / df_funnel['user_id'].shift(1) * 100, 2)

df_funnel = df_funnel.sort_values(by = ['cr(%)'], ascending = True)


plt.figure(figsize = (12, 8))

plt.barh(df_funnel['event_name'],
        df_funnel['cr(%)'])

plt.title('Классическая воронка (от первого шага)', fontsize = 14, fontweight = 'bold')
plt.xlabel('Конверсия (%)')
plt.ylabel('Шаги воронки')

for i, v in enumerate(df_funnel['cr(%)']):
    plt.text(v, i, str(v))
    
plt.tight_layout()

plt.show()
```

После регистрации и до онбординга теряется где-то 2–7% пользователей. Сильные потери начинаются после онбординга.

Построим воронку step by step

```python
plt.figure(figsize = (12, 8))

df_step_by_step = df_funnel.dropna(subset = ['step_by_step_cr']).copy()


plt.barh(df_step_by_step['event_name'],
        df_step_by_step['step_by_step_cr'])

plt.title('Воронка (step_by_step)', fontsize = 14, fontweight = 'bold')
plt.xlabel('Конверсия (%)')
plt.ylabel('Шаги')

for i, v in enumerate(df_step_by_step['step_by_step_cr']):
    plt.text(v, i, str(v))
    
plt.tight_layout()

plt.show()
```

После онбординга бизнес теряем много пользователей:

* После онбординга потери ~20% (относительно предыдущего шага);
* После первого депозита ~45%;
* После первой покупки активов ~65%.

**Влияние уровня риска актива на открытие второго депозита.**

```python
display(df_hist.head())
print()
print('----------------------------------------------------------------------------------------')
print()
display(df_hist['risk_level'].unique())
```

```python
df_first_purchase = df_hist[df_hist['event_name'] == 'asset_purchase'].copy()

df_second_deposit = df_hist[df_hist['event_name'] == 'second_deposit'].copy()
```

```python
df_risk = df_first_purchase.sort_values(by = 'event_ts').copy()
df_risk = df_risk.groupby('user_id')['risk_level'].first().reset_index()

# пользователей со вторым депозитом
second_deposit = set(df_second_deposit['user_id'].unique())

# определяем оформил ли пользователь 2-ой депозит
df_risk['2_deposit'] = df_risk['user_id'].apply(lambda x: 1 if x in second_deposit else 0)

# группируем по рискам и находим: кол-во первых покупок и кол-во оформленных 2-х депозитов
df_risk_group = df_risk.groupby('risk_level').agg(cnt_purchase = ('2_deposit', 'count'),
                                                 cnt_2_deposit = ('2_deposit', 'sum')).reset_index()
                                                  
df_risk_group['cr(%)'] = round(df_risk_group['cnt_2_deposit'] / df_risk_group['cnt_purchase'] * 100, 2)

df_risk_group.sort_values(by = 'cr(%)')
```

```python
plt.figure(figsize = (10, 6))

plt.bar(df_risk_group['risk_level'],
       df_risk_group['cr(%)'])

plt.title('Конверсия из 1-ой покупки в оформление 2-го депозита', fontsize = 14, fontweight = 'bold')
plt.xlabel('Уровень риска')
plt.ylabel('Конверсия')

for i, v in enumerate(df_risk_group['cr(%)']):
    plt.text(i, v, str(v), ha = 'center', va = 'bottom')
    
plt.tight_layout()

plt.show()
```

Самая низкая конверсия у активов с высокой степенью риска (26,36%), что может свидетельствовать о том, что клиенты сильно обжигаются на первых покупках и далее бояться совершать операции. 

У активов с риском `medium` - 42,97%. У `low` - 46,41%.

**Анализ ключевой метрики на исторических данных.**

```python
sum_deposits = df_hist['amount'].sum() # сумма всех депозитов

total_users = df_hist['user_id'].nunique() # всего пользователей

kpi = (sum_deposits / total_users).round(2)

print(f'Ключевая метрика (средняя сумма всех депозитов на одного пользователя ) на предпериоде = {kpi} у.е.')
```

```python
plot_hist = df_hist.groupby('user_id')['amount'].sum().reset_index()


plt.figure(figsize = (10, 6))
plt.hist(plot_hist['amount'],
        bins = 20,
        edgecolor = 'black')

avg = round(plot_hist['amount'].mean(), 2)

plt.axvline(avg,
           linestyle = '--',
            color = 'red',
           label = f'ARPU = {avg:.2f}')

plt.title('Распределение сумм депозитов', fontsize = 14, fontweight = 'bold')
plt.xlabel('сумма депозитов')
plt.ylabel('частота')

plt.legend()

plt.tight_layout()

plt.show()
```

**Промежуточные итоги:**

* Ежедневно в сервис приходит 620 - 710 пользователей;
* Наибольший отток начинается после онбординга, и с последующим шагом он увеличивается;
* Пользователи, которые изначально инвестируют в наиболее рискованные активы, менее склонны к повторному пополнению депозита. Чем меньше риск, тем выше конверсия из первой покупки в оформление второго депозита;
* На предпериоде средняя сумма депозитов на одного пользователя = 91,56 у.е.
* Гистограмма имеет скос вправо, что логично для показателей, которые отображают суммы покупок

**Влияние новой фичи на конверсию во второй депозит с учётом уровня риска купленного актива.**

Проанализируем, как категория риска купленного актива влияет на вероятность открытия второго депозита в тестовой и контрольной группах. 


```python
df_first_purchase_ab = df_abt[df_abt['event_name'] == 'asset_purchase'].copy()

df_second_deposit_ab = df_abt[df_abt['event_name'] == 'second_deposit'].copy()
```

```python
from scipy.stats import ttest_ind

df_risk_ab = df_first_purchase_ab.sort_values(by = 'event_ts').copy()
df_risk_ab = df_risk_ab.groupby('user_id')[['risk_level', 'group']].first().reset_index()

second_deposit_users = set(df_second_deposit_ab['user_id'].unique())

df_risk_ab['2nd_deposit'] = df_risk_ab['user_id'].apply(lambda x: 1 if x in second_deposit_users else 0)

alpha_bonf = 0.05 / 3

result = []

for risk in ['low', 'medium', 'high']:
    
    risk_data = df_risk_ab[df_risk_ab['risk_level'] == risk] # оставялем определенный риск
    
    control_data = risk_data[risk_data['group'] == 'control'] # группа контроля
    test_data = risk_data[risk_data['group'] == 'test'] # группа тест
    
        
    _, p_value = ttest_ind(test_data['2nd_deposit'], control_data['2nd_deposit'])
    
    result.append({
        'risk': risk,
        'cr_control_group': round(control_data['2nd_deposit'].mean() * 100, 2),
        'cr_test_group': round(test_data['2nd_deposit'].mean() * 100, 2),
        'lift': round(test_data['2nd_deposit'].mean()*100 - control_data['2nd_deposit'].mean() * 100, 2),
        'p_value': p_value.round(10),
        'alpha': alpha_bonf
    })

    
df_result = pd.DataFrame(result)
df_result['statistical_significance'] = df_result['p_value'] < alpha_bonf
df_result
```

Для всех видов риска конверсия в тестовой группе оказалась лучше, но не везде изменение статистически значимо:

* low: в тестовой группе метрика изменилась на 7,64, но изменение статистически не значимо;
* medium: изменение на +2,51, но также не стат. значимо;
* high: изменение на +18,34, изменение статистически значимо.

**Анализ метрик А/В-эксперимента.** 

Проанализируем такие метрики:
- Ключевая метрика - средняя сумма всех депозитов на одного пользователя (включая тех, кто установил приложение или открыл веб-версию).
- Барьерная метрика - конверсия из регистрации в первый депозит.
- Вспомогательная метрика 1 - конверсия из первого депозита во второй.
- Вспомогательная метрика 2 - средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит.

Анализ ключевой метрики - средняя сумма всех депозитов на одного пользователя:

```python
df_test = df_abt[df_abt['group'] == 'test'].copy()
df_control = df_abt[df_abt['group'] == 'control'].copy()

amount_test = round(df_test['amount'].sum(), 2)
amount_control = round(df_control['amount'].sum(), 2)

total_users_test = df_test['user_id'].nunique()
total_users_control = df_control['user_id'].nunique()

kpi_test = amount_test / total_users_test
kpi_control = amount_control / total_users_control

absolut_diff = round(kpi_test - kpi_control, 2)
relative_diff = round(((kpi_test - kpi_control) / kpi_control) * 100, 2)

df_kpi = pd.DataFrame({
    'метрика': ['сумма_депозитов', 'кол-во пользователей', 'kpi'],
    'контрольная группа': [amount_control, total_users_control, kpi_control],
    'тестовая группа': [amount_test, total_users_test, kpi_test],
    'абсолютное_изменение': [amount_test - amount_control, total_users_test - total_users_control, absolut_diff],
    'относительное_изменение_%': [(amount_test - amount_control) / amount_control * 100,
                                 (total_users_test - total_users_control) / total_users_control * 100,
                                 relative_diff]
})

display(df_kpi)
```

Проверим статистическую значимость различий ключевой метрики между группами эксперимента:

```python
alpha = 0.05

user_control = df_control.groupby('user_id')['amount'].sum().reset_index()
user_control.columns = ['user_id', 'total_deposit']

user_test = df_test.groupby('user_id')['amount'].sum().reset_index()
user_test.columns = ['user_id', 'total_deposit']

_, p_value = ttest_ind(user_control['total_deposit'], user_test['total_deposit'])

print(f' p_value = {p_value}')

if p_value < alpha:
    print('✅ изменение статистически значимое ✅')
else:
    print('❌ изменение НЕ является статистически значимым ❌')
```

Исследуем барьерную метрику - конверсия из регистрации в первый депозит.

```python
registration_users = set(df_abt[df_abt['event_name'] == 'registration']['user_id'].unique())

first_deposit_users = set(df_abt[df_abt['event_name'] == 'first_deposit']['user_id'].unique())

user_group = df_abt[['user_id', 'group']].drop_duplicates(subset = 'user_id')
user_group['registration'] = user_group['user_id'].isin(registration_users).astype('int')
user_group['first_deposit'] = user_group['user_id'].isin(first_deposit_users).astype('int')

df_cr = user_group[user_group['registration'] == 1].copy()
df_cr['cr'] = df_cr['first_deposit']

control_data = df_cr[df_cr['group'] == 'control']['cr']
test_data = df_cr[df_cr['group'] == 'test']['cr']

control_cr = control_data.mean() * 100
test_cr =  test_data.mean() * 100
diff = test_cr - control_cr
rel_diff = diff /  control_cr * 100

print(f'Контрольная группа: конверсия из регистрации в 1-й депозит - {control_cr:.2f} %')
print(f'Тестовая группа: конверсия из регистрации в 1-й депозит - {test_cr:.2f} %')
print(f'абсолютная разница между группами = {diff:.2f}')
print(f'относительная разница = {rel_diff:.2f}%')

print()

_, p_value = ttest_ind(control_data, test_data)
print(f'p-value = {p_value}')
if p_value < 0.05:
    print('✅ изменение статистически значимое ✅')
else:
    print('❌ изменение НЕ является статистически значимым ❌')
```

Анализ вспомогательной метрики 1 - конверсия из первого депозита во второй. Для разнообразия используем z-тест пропорций

```python
from statsmodels.stats.proportion import proportions_ztest

first_deposit_users = set(df_abt[df_abt['event_name'] == 'first_deposit']['user_id'].unique())
second_deposit_users = set(df_abt[df_abt['event_name'] == 'second_deposit']['user_id'].unique())

group = df_abt[['user_id', 'group']].drop_duplicates(subset = 'user_id')

group['1_deposit'] = group['user_id'].isin(first_deposit_users).astype('int')
group['2_deposit'] = group['user_id'].isin(second_deposit_users).astype('int')

df_1_2 = group[group['1_deposit'] == 1].copy()
df_1_2['cr'] = df_1_2['2_deposit']

control = df_1_2[df_1_2['group'] == 'control']['cr']
test = df_1_2[df_1_2['group'] == 'test']['cr']

cr_control = control.mean() * 100
cr_test = test.mean() * 100
differ = cr_test - cr_control
rel_differ = differ / cr_control * 100

print(f'Контрольная группа: конверсия из первого депозита во второй - {cr_control:.2f} %')
print(f'Тестовая группа: конверсия из первого депозита во второй - {cr_test:.2f} %')
print(f'абсолютная разница между группами = {differ:.2f}')
print(f'относительная разница = {rel_differ:.2f}%')

print()

control_success = control.sum()
control_total = len(control)
test_success = test.sum()
test_total = len(test)

m = [test_success, control_success] # успех
n = [test_total, control_total] # размер

_, p_valie_ztest = proportions_ztest(m, n, alternative = 'two-sided') 

print(f'p-value = {p_valie_ztest}')
if p_valie_ztest< 0.05:
     print('✅ изменение статистически значимое ✅')
else:
    print('❌ изменение НЕ является статистически значимым ❌')
```

Анализ вспомогательной метрики 2 - средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит:

```python
deposits = ['first_deposit', 'second_deposit']

users_deposit = df_abt[df_abt['event_name'].isin(deposits)].copy()
users_deposit = users_deposit.groupby(['user_id', 'group'])['amount'].sum().reset_index()

control = users_deposit[users_deposit['group'] == 'control']['amount']
test = users_deposit[users_deposit['group'] == 'test']['amount']

control_mean = control.mean()
test_mean = test.mean()
difference = test_mean - control_mean
rel_difference = difference / control_mean * 100

print(f'Контрольная группа: ARPPU = {control_mean:.2f}')
print(f'Тестовая группа: ARPPU = {test_mean:.2f}')
print(f'Абсолютная разница = {difference:.2f}')
print(f'Относительная разница = {rel_difference:.2f} %')

print()

p_value = ttest_ind(test, control).pvalue

print(f'p-value = {p_value}')
if p_value < 0.05:
    print('✅ изменение статистически значимое ✅')
else:
    print('❌ изменение НЕ является статистически значимым ❌')

```

Накопленная динамика изменения KPI по дням эксперимента для каждой группы

```python
df_abt['event_dt'] = df_abt['event_ts'].dt.date
df_abt['event_dt'] = pd.to_datetime(df_abt['event_dt'])

records = []

for date in sorted(df_abt['event_dt'].unique()):
    df_curr_date = df_abt[df_abt['event_dt'] <= date]
    
    df_curr_group = df_curr_date.groupby('group').agg(total_amount = ('amount', 'sum'),
                                                     cnt_users = ('user_id', 'nunique')).reset_index()
    
    result = df_curr_group
    result['arpu'] = result['total_amount'] / result['cnt_users']
    result['date'] = date
    records.append(result)
    
df_arpu = pd.concat(records)
#df_arpu

plt.figure(figsize = (12, 8))

for group in df_arpu['group'].unique():
    df_plot = df_arpu[df_arpu['group'] == group]
    
    plt.plot(df_plot['date'], df_plot['arpu'], label = group, marker = 'o')
    
plt.title('Накопленная динамика изменения ключевой метрики (ARPU)', fontsize = 14, fontweight = 'bold')
plt.xlabel('Дни')
plt.ylabel('ARPU')

plt.xticks(df_plot['date'], rotation = 45)

plt.legend()
plt.tight_layout()
    
plt.show()
```

Стабильность p-value во время эксперимента:

```python

pval_records = []

for date in sorted(df_abt['event_dt'].unique()):
    
    df_pval = df_abt[df_abt['event_dt'] <= date]
    
    df_pval_group = df_pval.groupby(['user_id', 'group'])['amount'].sum().reset_index()
    
    control = df_pval_group[df_pval_group['group'] == 'control']['amount']
    test = df_pval_group[df_pval_group['group'] == 'test']['amount']
    
    p_value = ttest_ind(control, test, equal_var = False).pvalue
    
    pval_records.append({
        'date': date,
        'p_value': p_value
    })
    

df_val = pd.DataFrame(pval_records)

# df_val

plt.figure(figsize = (12, 8))

plt.plot(df_val['date'], df_val['p_value'], marker = 'o')

plt.axhline(0.05, color = 'black', linewidth = 1, label = f'alpha 0.05')

plt.title('Динамика p-value по дням эксперемента', fontsize = 14, fontweight = 'bold')
plt.xlabel('дни')
plt.ylabel('p-value')

plt.legend()
plt.xticks(df_val['date'], rotation = 45)

plt.tight_layout()

plt.show()
```

### 4. Анализ изменений суммы депозитов на платящего пользователя

Новая фича могла повлиять на поведение пользователей.
- Пользователи, которые раньше вносили небольшие суммы, могли стать более осторожными, сократить свои вложения или совсем перестать платить. Это отразится в снижении 25-го перцентиля суммы депозитов в тестовой группе.
- Пользователи, которые склонны к более крупным инвестициям, могли сильнее вовлечься в продукт и начать вносить больше средств. Это отразится в росте 75-го перцентиля.

Используя бутстрап, сравним разницы перцентилей (25, 50, 75) суммы всех депозитов на платящего пользователя в контрольной и тестовой группах

```python
import numpy as np

np.random.seed(42)

deposits = ['first_deposit', 'second_deposit']

users_payers = df_abt[df_abt['event_name'].isin(deposits)]
users_payers = users_payers.groupby(['user_id', 'group'], as_index = False)['amount'].sum()
# users_payers

control_data = users_payers[users_payers['group'] == 'control']['amount']
test_data = users_payers[users_payers['group'] == 'test']['amount']

p25_diff = []
p50_diff = []
p75_diff = []


iterations = 1001

for _ in range(iterations):
    
    boot_control = np.random.choice(control_data, len(control_data), replace = True)
    boot_test = np.random.choice(test_data, len(test_data), replace = True)
    
    # расчет 25 перцентиля
    control_25p = np.percentile(boot_control, 25)
    test_25p = np.percentile(boot_test, 25)
    p25_diff.append(test_25p - control_25p)
    
    # расчет 50 перцентиля
    control_50p = np.percentile(boot_control, 50)
    test_50p = np.percentile(boot_test, 50)
    p50_diff.append(test_50p - control_50p)
    
    # расчет 75 перцентиля
    control_75p = np.percentile(boot_control, 75)
    test_75p = np.percentile(boot_test, 75)
    p75_diff.append(test_75p - control_75p)
    
    

p25_diff = np.array(p25_diff)
p50_diff = np.array(p50_diff)
p75_diff = np.array(p75_diff)

#p25_diff_ci = np.percentile(p25_diff, [2.5, 97.5])
#print(np.percentile(control_data, 25))
#print(np.percentile(test_data, 25))
#print(p25_diff_ci[0], p25_diff_ci[1])
```

```python
def func_boot_and_hist(boot_diff, control, test, pers):
  
    '''
    функция на вход принимает:

    - массив разниц перцентилей тестовой и контрольной групп;
    - данные из контрольной группы;
    - данные из тестовой группы;
    - перцентиль (25, 50, 75).

    Считает 95% доверительный интервал.

    Выводит информацию:

    1. значение (перцентиль) в контрольной группе;
    2. значение (перцентиль) в тестовой группе;
    3. доверительный интервал;
    4. строит гистограмму.
    '''
    
    diff_ci = np.percentile(boot_diff, [2.5, 97.5]) # считаем 95% доверительный интервал
    
    print('--' * 60)
    print(f'{pers}-й перцентиль контрольной группы: {np.percentile(control, pers)}')
    print(f'{pers}-й перцентиль тестовой группы: {np.percentile(test, pers)}')
    print(f'95-й доверительный интервал разности {pers}-х перцентилей: [{diff_ci[0]:.2f}, {diff_ci[1]:.2f}]')
    print('--' * 60)
    
    print()
    
    
    plt.figure(figsize = (12, 8))
    
    plt.hist(boot_diff,
            bins= 25,
            edgecolor = 'black')
    
    plt.axvline(diff_ci[0],
               linestyle = '-',
               color = 'red',
               label = f'нижняя граница ДИ: {diff_ci[0]}')
    
    plt.axvline(diff_ci[1],
               linestyle = '-',
               color = 'red',
               label = f'верхняя граница ДИ: {diff_ci[1]}')
    
    plt.title(f'Бутстрап распределение: разность {pers}-го перцентиля', fontsize = 14, fontweight = 'bold')
    
    plt.legend()
    plt.tight_layout()
    
    plt.show()
```

```python
func_boot_and_hist(p25_diff, control_data, test_data, 25)
```

Гипотеза о том, что пользователи, которые вносили небольшие суммы, стали осторожнее, подтвердилась. Пользователи из тестовой группы пополняют депозиты на 44–50 у. е. меньше, чем в контрольной.

```python
func_boot_and_hist(p50_diff, control_data, test_data, 50)
```

Аналогичная ситуация и с пользователями, которые тратят среднюю сумму, в тестовой группе пользователи пополняют депозиты на меньшую сумму по сравнению с контрольной группой.

```python
func_boot_and_hist(p75_diff, control_data, test_data, 75)
```

Ситуация с пользователями, которые тратят суммы побольше остальных, противоположная. Тут тестовая группа инвестирует больше, чем контрольная.

Гипотеза подтверждается.
